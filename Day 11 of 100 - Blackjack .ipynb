```python
## Import Libraries

import random
```

```python
## Set Global Variables

player_hand = []
player_score = -1 # Must start at -1 or else game immediately ends
dealer_hand = []
dealer_score = -1 # Must start at -1 or else game immediately ends

deck = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]
ace = 11

dealer_turn = True # How to end the game
player_turn = True # How to end the game
```

```python
## Functions for the game

def draw_card():
    # returns a card at random from deck list
    card = random.choice(deck)
    return card

def calculate_score(hand):
    # checking for a Blackjack
    if sum(hand) == 21 and len(hand) == 2:
        return 0
    elif sum(hand) > 21 and ace in hand:
        hand.remove(ace)
        hand.append(1)
    return sum(hand)


def compare():
    # comparing scores against each other to determine the winner
    if player_score == dealer_score:
        print(f"Player: {player_score} and dealer: {dealer_score} - it's a draw!")
    elif dealer_score == 21:
        print(f"Dealer wins with {dealer_score} Blackjack!")
    elif player_score == 21:
        print(f"Player wins with {player_score} Blackjack!")
    elif player_score > 21:
        print(f"You got {player_score} - You lose, dealer wins")
    elif dealer_score > 21:
        print(f"Dealer got {dealer_score} - dealer loses, you win")
    else:
        highest = max(player_score, dealer_score)
        print(f"winning score: {highest} ")
```

```python
## The Game

for i in range(2): # drawing 2 cards for both player and dealer
    player_hand.append(draw_card())
    dealer_hand.append(draw_card())
    
while player_turn:
    player_score = calculate_score(player_hand)
    dealer_score = calculate_score(dealer_hand)

    print(f"Your cards are: {player_hand}, Card total: {player_score}",
          f"Dealers card: {dealer_hand[0]}")
    
    if player_score == 0 or dealer_score == 0 or player_score > 21:
        player_turn = False
        dealer_turn = False
    else:
        deal_again = input("Do you want to draw another card? ").lower() # prompt user for another card
        if deal_again == "yes":
            player_hand.append(draw_card())
            calculate_score(player_hand) 
        else:
            player_turn = False


while dealer_turn:

    player_score = calculate_score(player_hand)  # Calls the function and stores the output into a variable
    dealer_score = calculate_score(dealer_hand)

    if dealer_score < 17:
        dealer_hand.append(draw_card())
        calculate_score(dealer_hand)
        if player_score == 0 or dealer_score == 0 or dealer_score > 21:
            dealer_turn = False

    else:
        dealer_turn = False


print(f"Your final hand: {player_hand}, Final Score: {player_score}")
print(f"Dealer final hand: {dealer_hand}, Final score {dealer_score}")
compare()
```
